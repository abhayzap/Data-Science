## 1 — Basic Function

```python
def hello():
    print("Hello Python")

hello()
```

## 2 — Function with Parameter

```python
def hello(name):
    print(f"Hello {name}")

hello("Abhay")
hello("Python")
```

## 3 — Function with Return (Sum of Two Numbers)

```python
def add(a, b):
    return a + b

print(add(5, 6))
print(add(5, 5))
print(add(9, 6))
```

## 4 — Average of Three Numbers

```python
def avg(a, b, c):
    return round((a + b + c) / 3, 2)

print(avg(5, 5, 5))
print(avg(5, 6, 9))
```

## 5 — Default Parameter

```python
def add(a, b=1):
    return a + b

print(add(5))     # uses default b=1
print(add(5, 5))  # overrides default
```

## 6 — Lambda Function

```python
# Square using lambda
square = lambda x: x * x
print(square(5))

# Add using lambda
add = lambda a, b: a + b
print(add(3, 7))

# Average using lambda
average = lambda a, b: (a + b) / 2
print(average(4, 6))
```

## 7 — Factorial Calculator

```python
def cal_fact(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

n = int(input("Enter a number: "))
print(f"Factorial of {n} is: {cal_fact(n)}")
```
