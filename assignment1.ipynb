## Problem 7 — Operator Precedence

```python
x = 10 + 3 * 2 ** 2
print(f"Result: {x}")  # Output: 22

a = (10 + 3) * 2 ** 2
b = 10 + (3 * 2) ** 2
c = (10 + 3 * 2) ** 2

print(a)  # 52
print(b)  # 46
print(c)  # 256
```

## Problem 11 — Integer & Fractional Parts

## Problem 12 — Variable Swap

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print(f"Before swap: a = {a}, b = {b}")

# Method 1 - Traditional (using temp variable)
temp = a
a = b
b = temp
print(f"After swap (Method 1): a = {a}, b = {b}")

# Method 2 - Pythonic one-liner
a, b = b, a
print(f"After swap (Method 2): a = {a}, b = {b}")
```
