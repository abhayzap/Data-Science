## 1 — Set Basics

```python
s = {1, 2, 2, 2, 2, 2}
print(s)        # {1, 2} - duplicates removed!
print(type(s))  # <class 'set'>

# Empty set
empty_set = set()  # NOT {} which creates empty dict
```

## 2 — Set Methods

```python
s = {1, 2, 3}
s.add(5)       # add element
print(s)

s.discard(10)  # safe remove - no error if missing
s.remove(1)    # remove - ERROR if not found
print(s)

s.pop()        # removes random element!
print(s)
```

## 3 — Set Operations

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 9}

print(s1.union(s2))         # all elements combined
print(s1.intersection(s2))  # common elements only
print(s1.difference(s2))    # in s1 but not s2
```

## 4 — Practice Problem

```python
# Given student-course data:
# 1. List all unique courses
# 2. List students enrolled in English
# 3. Create dictionary (student: set of courses)

info = {
    ("Alice", "Maths"),
    ("Bob", "Science"),
    ("Alice", "Science"),
    ("Charlie", "Math"),
    ("Bob", "Math"),
    ("Alice", "English"),
    ("Charlie", "English"),
}

# 1. Unique courses
course_set = set()
for tup in info:
    course_set.add(tup[1])
print(f"Unique courses: {course_set}")

# 2. Students in English
print("\nStudents in English:")
for name, course in info:
    if course == "English":
        print(name)

# 3. Dictionary of courses per student
student_courses = {}
for name, course in info:
    if student_courses.get(name) == None:
        student_courses.update({name: set()})
        student_courses[name].add(course)
    else:
        student_courses[name].add(course)
print(f"\nStudent courses: {student_courses}")
```
