## 1 — Print Hello World 5 Times (While Loop)

```python
i = 1
while i <= 5:
    print(f"hello world {i}")
    i += 1
```

## 2 — Print Numbers 1 to 5 (Forward & Reverse)

```python
# Forward
i = 1
while i <= 5:
    print(i)
    i += 1

print("---")

# Reverse
i = 5
while i >= 1:
    print(i)
    i -= 1
```

## 3 — Multiplication Table

```python
n = int(input("Enter a number: "))
i = 1
while i <= 10:
    print(f"{n} * {i} = {n*i}")
    i += 1
```

## 4 — Break Keyword

```python
i = 1
while i <= 10:
    if i == 6:
        break
    print(i)
    i += 1
print("Outside the loop")
```

## 5 — Continue Keyword (Skip Multiples of 3)

```python
i = 0
while i < 10:
    i += 1
    if i % 3 == 0:
        continue
    print(i)
```

## 6 — For Loop Basics

```python
# Iterate over a string
string = "hello"
for char in string:
    print(char)

print("---")

# Range variations
for i in range(5):        # 0 to 4
    print(i)

for i in range(1, 6):     # 1 to 5
    print(i)

for i in range(1, 10, 2): # 1,3,5,7,9
    print(i)

for i in range(10, 0, -1): # 10 to 1
    print(i)
```

## 7 — Count Letter 'i' in a Word

```python
word = "artificial intelligence"
count = 0
for ch in word:
    if ch == 'i':
        count += 1
print(f"Number of i's: {count}")
```

## 8 — Vowel Counter

```python
word = "artificial"
count = 0
for char in word:
    if char in "aeiou":
        count += 1
print(f"Number of vowels in '{word}': {count}")
```

## 9 — Odd Numbers Using Continue

```python
i = 1
while i <= 10:
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i += 1
```

## 10 — Sum of First N Natural Numbers

```python
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print(f"Sum of first {n} natural numbers: {total}")
```
