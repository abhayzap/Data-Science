## 1 — List Basics

```python
marks = [99, 89, 100, 65, 92]
print(marks)
print(marks[1])   # 89
print(len(marks)) # 5
```

## 2 — List Mutability

```python
marks = [99, 89, 100, 65, 92]
marks[2] = 70  # lists are mutable!
print(marks)   # [99, 89, 70, 65, 92]
```

## 3 — Mixed Data Types

```python
marks = [99, 89, 100, 65, 92, "abc", 101.01]
print(type(marks))  # <class 'list'>
```

## 4 — List Slicing

```python
marks = [99, 89, 100, 65, 92, "abc", 101.01]
print(marks[0:5])   # first 5 items
print(marks[5:])    # from index 5 to end
print(marks[::-1])  # reversed
```

## 5 — List Methods

```python
nums = [1, 2, 3]

nums.append(4)      # add to end
print(nums)         # [1, 2, 3, 4]

nums.insert(2, 10)  # insert at index 2
print(nums)         # [1, 2, 10, 3, 4]

nums.sort()         # sort ascending
print(nums)         # [1, 2, 3, 4, 10]

nums.reverse()      # reverse
print(nums)         # [10, 4, 3, 2, 1]
```

## 6 — Searching in a List

```python
nums = [1, 2, 3, 10, 4]
x = 10
idx = 0

for val in nums:
    if val == x:
        print(f"x found at idx={idx}")
        break
    idx += 1
```
