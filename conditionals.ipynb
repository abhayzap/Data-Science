## 1 — Age Category Classifier

```python
age = int(input("Enter your age: "))

if age < 13:
    print("You are a child")
elif 13 <= age < 18:
    print("You are a teen")
elif 18 <= age < 65:
    print("You are an adult")
else:
    print("You are a senior citizen")
```

## 2 — Multiple of 5 Checker

```python
n = int(input("Enter a number: "))

if n % 5 == 0:
    print(f"{n} is a multiple of 5")
else:
    print(f"{n} is not a multiple of 5")
```

## 3 — Even or Odd Checker

```python
n = int(input("Enter a number: "))

if n % 2 == 0:
    print(f"{n} is an even number")
else:
    print(f"{n} is an odd number")
```

## 4 — Login System (Nested Conditions)

```python
username = input("Enter your username: ")
password = input("Enter your password: ")

if username == "admin" and password == "pass":
    print("Login successful")
else:
    if username != "admin":
        print("Invalid username")
    else:
        print("Invalid password")
```

## 5 — Traffic Light (Match-Case)

```python
color = input("Enter a color (red/green/yellow): ")

match color:
    case "red":
        print("Stop")
    case "green":
        print("Go")
    case "yellow":
        print("Slow down")
    case _:
        print("Invalid color")
```
