## 1 — Tuple Basics

```python
tup = (1, 2, 3, 4, 5)
print(tup)
print(type(tup))  # <class 'tuple'>
print(len(tup))   # 5
```

## 2 — Tuple Immutability

```python
# Lists are mutable
nums = [1, 2, 3]
nums[0] = 99  # works!

# Tuples are immutable
tup = (1, 2, 3)
# tup[0] = 99  # TypeError! Cannot modify tuple
```

## 3 — Tuple Slicing

```python
tup = (1, 2, 3, 4, 5)
print(tup[-1])    # 5
print(tup[1:4])   # (2, 3, 4)
print(tup[::2])   # (1, 3, 5)
print(tup[::-1])  # (5, 4, 3, 2, 1)
```

## 4 — Tuple Methods

```python
tup = (1, 2, 3, 2, 4, 2, 5)
print(tup.index(2))  # 1 (first occurrence)
print(tup.count(2))  # 3 (total count)
```

## 5 — Sum of Tuple

```python
tup = (1, 2, 3, 4, 5)
total = 0
for val in tup:
    total += val
print(f"Sum of the tuple: {total}")  # 15

# Built-in shortcut
print(sum(tup))  # 15
print(max(tup))  # 5
print(min(tup))  # 1
```
