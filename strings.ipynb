## 1 — String Length

```python
word = "python"
print(len(word))  # 6
```

## 2 — String Concatenation

```python
word1 = "I love"
word2 = "python"

# Method 1 - + operator
print(word1 + " " + word2)

# Method 2 - f-string
sentence = word1 + " " + word2
print(f"{sentence}")
```

## 3 — String Indexing

```python
word = "python"

# Positive indexing
print(word[0])   # p
print(word[1])   # y
print(word[2])   # t

# Negative indexing
print(word[-1])  # n (last)
print(word[-2])  # o (second last)
```

## 4 — String Slicing

```python
word = "python"

print(word[0:3])   # pyt
print(word[2:5])   # tho
print(word[1:4])   # yth
print(word[:4])    # pyth
print(word[3:])    # hon
print(word[::2])   # pto
print(word[::-1])  # nohtyp (reverse!)
```

## 5 — String Iteration

```python
word = "python"
for ch in word:
    print(ch)
```

## 6 — `in` Operator

```python
string = "hello"
if 'o' in string:
    print("o exists in the string")
```

## 7 — String Formatting

```python
a = 5
b = 10

# .format() method
print("language is {}".format("python"))
print("sum of {} & {} is {}".format(a, b, a+b))
print("sum of {1} & {0} is {2}".format(a, b, a+b))  # index based
print("values of vars {a} & {b}".format(a=5, b=10))  # value based

# f-string (preferred)
print(f"sum of {a} & {b} is {a+b}")
print(f"avg of {a} & {b} is {(a+b)/2}")
```
