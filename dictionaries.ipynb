## 1 — Dictionary Basics

```python
info = {
    "name": "abhay",
    "cgpa": 9.2,
    "subject": ["maths", "science"],
    3.14: "PI"
}

info["cgpa"] = 9.6  # update value
print(info["cgpa"])  # 9.6
```

## 2 — Dictionary Methods

```python
info = {
    "name": "abhay",
    "cgpa": 9.2,
    "subject": ["maths", "science"]
}

print(list(info.keys()))    # all keys
print(list(info.values()))  # all values
print(info.items())         # key-value pairs

print(info.get("cgpa"))    # 9.2
print(info.get("cgpa2"))   # None - safe!
print(info.get("cgpa2", "Not found"))  # default value

info.update({"city": "Calgary"})
print(info)
```
